```python
import os
import pandas as pd
import numpy as  np
import librosa
from librosa import feature
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
fn_list_i = [
 feature.chroma_stft,
 feature.spectral_centroid,
 feature.spectral_bandwidth,
 feature.spectral_rolloff
]
 
fn_list_ii = [
 feature.rms,
 feature.zero_crossing_rate
]

def get_feature_vector(y,sr): 
   feat_vect_i = [ np.mean(funct(y=y, sr=sr)) for funct in fn_list_i]
   feat_vect_ii = [ np.mean(funct(y=y)) for funct in fn_list_ii] 
   feature_vector = feat_vect_i + feat_vect_ii 
   return feature_vector
```

```python
df = pd.DataFrame()
```

```python
for each_actor in os.listdir('dataset/'):
    for each_file in os.listdir(os.path.join('dataset', each_actor)):
        
        emotion = each_file.split("-")[2]
        
        y , sr = librosa.load(os.path.join('dataset', each_actor, each_file), sr=None)
        feature_vector = get_feature_vector(y, sr)
        df = df.append([feature_vector + [str(emotion)]])
```
